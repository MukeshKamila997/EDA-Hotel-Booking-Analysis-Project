**Which column has the most missing data?**

```python
max_null_value = df.isnull().sum()
max_null_value
```

```python
# lets Drop the columns with maximum null values

df = df.drop(['company'], axis = 1)
```

```python
# Replace the null values of “children” column by 0

df['children'].replace(np.nan,0,inplace=True)
df['children'].unique()
```

```python
# Replace the null values in the “agent” column by mean of that column
avg_agent = df['agent'].astype('float').mean(axis=0)
# df['agent'].fillna(avg_agent,inplace=True)
df['agent'].replace(np.nan,avg_agent,inplace=True)
```

```python
avg_agent
```

## Hotel bookings and cancelation 

```python
bookings = df.groupby(['hotel'])['is_canceled'].value_counts().unstack()
bookings
```

```python
bookings.plot(kind='bar',edgecolor = 'black',figsize=(10,5))
plt.title('Hotel bookings and cancelation',fontsize = 20)
plt.xlabel('Hotels',fontsize = 15)
plt.ylabel('Bookings and Cancelation',fontsize = 15)
```

```python
# '''Reason behind cancelation can be cheaper price in competitive market. ''' 
```

## Year-wise Bookings

```python
year_and_month = df.groupby(['arrival_date_year'])['arrival_date_month'].value_counts().unstack()
year_and_month
```

```python
year_and_month.plot(kind = 'bar', figsize = (11,6),edgecolor = 'black',width = 0.6,linewidth = 1.2)
plt.title('Total Bookings',fontsize = 20)
plt.xlabel('arrival_date_year',fontsize = 14)
```

```python
# Monthwise 2017 has less booking then 2016, reason can be price, new competitors, offers & discounts.
```

## Boom in Bookings 

```python
df['arrival_date_month'].value_counts()
```

```python
plt.figure(figsize=(14,8))
sns.set(style="darkgrid")

ax = sns.countplot(df['arrival_date_month'], order = df['arrival_date_month'].value_counts().sort_values(ascending = True).index)
ax.set(title='Boom in Bookings')
```

```python
# so in January least number of guests arrived compare to other months, but August has the most arrival rate.  
```

```python
# Analysis for August Bookings
day = df[df['arrival_date_month'] == 'August']['arrival_date_day_of_month'].value_counts()
day
```

```python
plt.figure(figsize = (14,8))
sns.countplot(df[df['arrival_date_month']== 'August']['arrival_date_day_of_month'], order = (day).index)
plt.title('August Bookings',fontsize = 20)
```

```python
# Around 14-18 August most arrivals are there, so the hotel management can manage or arrange stocks and staffs as per that.
```

## Hotel Preference By Customer type

```python
hotel_preference = df.groupby(['customer_type'])['hotel'].value_counts().unstack()
hotel_preference
```

```python
# hotel_preference.plot(kind='bar',edgecolor='black',figsize = (12,6))
plt.figure(figsize=(12,8))
sns.countplot(x = 'customer_type',hue ='hotel',data=df)
plt.title('Hotel Preference',fontsize=20)
plt.xlabel('Customer_type',fontsize=15)
```

## Night Stays preference(Weekend)

```python
nights = df[df['stays_in_weekend_nights'] != 0]['stays_in_weekend_nights'].value_counts()
nights
```

```python
plt.figure(figsize=(13,7))
sns.lineplot(data = nights)
plt.title('Night_stays_weekend',fontsize=20)
plt.xlabel('Weekend_Numbers',fontsize=15)
```

```python
# Most preferably, guests would like to book or stay for 1-2 weekend nights in hotels.
```

## Night Stays preference(Weekday)

```python
weekday = df[df['stays_in_week_nights']!=0]['stays_in_week_nights'].value_counts()
weekday
```

```python
plt.figure(figsize=(13,7))
sns.countplot(df[df['stays_in_week_nights']!=0]['stays_in_week_nights'],facecolor='green')
plt.title('Night_Stays_Weekday',fontsize=20)
plt.xlabel('Weekday_Numbers',fontsize=15)
```

## The average number of nights spent.

```python
df['nights_spent'] = df.apply(lambda x: x['stays_in_weekend_nights'] + x['stays_in_week_nights'],axis=1)
df['nights_spent'].mean().round(2)
```

```python
# So the average nights spent per guest is approxmately 3 nights.
```

```python
plt.figure(figsize = (13,7))
sns.distplot(df['nights_spent'],hist=True)
plt.title('Nights_spent',fontsize = 20)
```

```python
# Above distribution plot shows maximum of 1-3 nights spent.
```
